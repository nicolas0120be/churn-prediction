```python
# notebooks/03_model_training_and_evaluation.ipynb

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib # For saving/loading models
import sys
import os

# Add the src directory to the Python path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '../src')))

from data_processor import DataProcessor
from model_trainer import ModelTrainer

# Configure matplotlib and seaborn for better visuals
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12

# --- Section 1: Load Processed Data ---
print("--- Section 1: Loading Processed Data ---")

# Re-initialize DataProcessor and get the processed data.
# In a real scenario, you might load the saved npy files from the previous notebook.
# For this structured notebook flow, we'll re-run the processor to ensure consistency.
# Initialize DataProcessor
processor = DataProcessor(data_path=r"C:\Users\nicol\OneDrive\Documentos\customer-churn-prediction\data\WA_Fn-UseC_-Telco-Customer-Churn.csv")
processor.load_data()
processor.clean_data() # A
processor.preprocess_data()
X_train_processed, X_test_processed, y_train, y_test = processor.split_data(
    test_size=0.2, random_state=42, apply_smote=True
)

if X_train_processed is None:
    raise ValueError("Failed to load or preprocess data. Please check previous notebook.")

print(f"Loaded X_train_processed shape: {X_train_processed.shape}")
print(f"Loaded y_train shape: {y_train.shape}")
print(f"Loaded X_test_processed shape: {X_test_processed.shape}")
print(f"Loaded y_test shape: {y_test.shape}")

# --- Section 2: Initialize Model Trainer and Define Models ---
print("\n--- Section 2: Initializing Model Trainer and Defining Models ---")

trainer = ModelTrainer(random_state=42)
trainer.define_models()

# Dictionary to store all trained models for comparison
trained_models = {}

# --- Section 3: Train and Evaluate Models ---
print("\n--- Section 3: Training and Evaluating Models ---")

# Iterate through each defined model, train, and evaluate
for name, model_info in trainer.models.items():
    print(f"\n--- Training and Evaluating {name} ---")
    # For quick runs during development, you can set `param_search_type='none'`
    # or reduce `n_iter_random` for RandomizedSearchCV.
    # For the final run, 'grid' or 'random' with sufficient iterations is recommended.
    current_model = trainer.train_model(name, X_train_processed, y_train,
                                        param_search_type='random', n_iter_random=20) # Reduced for quicker demo
    if current_model:
        trained_models[name] = current_model
        metrics = trainer.evaluate_model(current_model, X_test_processed, y_test, model_name=name)
        trainer.plot_confusion_matrix(current_model, X_test_processed, y_test, model_name=name)

# --- Section 4: Model Comparison ---
print("\n--- Section 4: Model Comparison ---")

# Display the performance metrics for all trained models
print("\nSummary of all model metrics:")
if hasattr(trainer, 'metrics') and not trainer.metrics.empty:
    display(trainer.metrics.sort_values(by='ROC-AUC', ascending=False))
else:
    print("No metrics to display. Please check if models were trained successfully.")

# Plot ROC curves for all models for visual comparison
if trained_models:
    trainer.plot_roc_curve(trained_models, X_test_processed, y_test)
    trainer.plot_precision_recall_curve(trained_models, X_test_processed, y_test)

# --- Section 5: Save All Trained Models ---
print("\n--- Section 5: Saving All Trained Models ---")

models_dir = os.path.join('..', 'models')
os.makedirs(models_dir, exist_ok=True)

for model_name, model in trained_models.items():
    model_file = f"{model_name.replace(' ', '_').lower()}_churn_model.pkl"
    model_save_path = os.path.join(models_dir, model_file)
    trainer.save_model(model, model_save_path)
    print(f"Model '{model_name}' saved to: {model_save_path}")

# Optionally, keep the best model logic below if you want to highlight the best one
best_model_summary = trainer.get_best_model_summary()

if best_model_summary is not None:
    best_model_name = best_model_summary['Model']
    best_model = trained_models[best_model_name]
    trainer.best_model = best_model

    print(f"\nThe best model is: {best_model_name}")

    # Optional: Load the model back to verify
    best_model_file = f"{best_model_name.replace(' ', '_').lower()}_churn_model.pkl"
    best_model_path = os.path.join(models_dir, best_model_file)
    loaded_best_model = trainer.load_model(best_model_path)
    if loaded_best_model:
        print("\nVerification: Evaluating the loaded best model:")
        trainer.evaluate_model(loaded_best_model, X_test_processed, y_test, model_name=f"Loaded {best_model_name}")

print("\nAll models have been saved. Model Training and Evaluation Complete. Proceed to insights and recommendations.")
```
