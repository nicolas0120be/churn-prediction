```python
# notebooks/01_data_cleaning_and_eda.ipynb

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os

# Add the src directory to the Python path to import custom modules
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '../src')))

# --- Section 1: Load Data ---
print("--- Section 1: Loading Data ---")

# Load the dataset directly using pandas (fix: remove broken DataProcessor import and usage)
data_path = r"C:\Users\nicol\OneDrive\Documentos\customer-churn-prediction\data\WA_Fn-UseC_-Telco-Customer-Churn.csv"
df = pd.read_csv(data_path)

# Get the raw DataFrame for initial inspection
if df is not None:
    print("\nFirst 5 rows of the dataset:")
    display(df.head())
    print(f"\nDataset shape: {df.shape}")
    print("\nDataset information:")
    df.info()
    print("\nMissing values before cleaning:")
    print(df.isnull().sum())

# --- Section 2: Data Cleaning ---
print("\n--- Section 2: Data Cleaning ---")
# Example cleaning: Remove spaces, convert TotalCharges to numeric, drop missing
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df = df.dropna()
df = df.reset_index(drop=True)
df['Churn'] = df['Churn'].map({'No': 0, 'Yes': 1})

df_cleaned = df

if df_cleaned is not None:
    print("\nMissing values after cleaning:")
    print(df_cleaned.isnull().sum())
    print("\nData types after cleaning:")
    print(df_cleaned.dtypes)
    print("\nValue counts for 'Churn' target variable:")
    print(df_cleaned['Churn'].value_counts())
    print(f"Churn rate: {df_cleaned['Churn'].mean():.2%}")

# --- Section 3: Exploratory Data Analysis (EDA) ---
print("\n--- Section 3: Exploratory Data Analysis (EDA) ---")

# Target variable distribution
plt.figure(figsize=(6, 5))
sns.countplot(x='Churn', data=df_cleaned)
plt.title('Distribution of Churn (0: No Churn, 1: Churn)')
plt.xlabel('Churn')
plt.ylabel('Number of Customers')
plt.show()

# Categorical Features Analysis
categorical_cols = df_cleaned.select_dtypes(include='object').columns.tolist()
# Exclude 'customerID' if it was kept initially or is still present
if 'customerID' in categorical_cols:
    categorical_cols.remove('customerID')

print("\nAnalyzing categorical features against Churn:")
for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df_cleaned, x=col, hue='Churn', palette='viridis')
    plt.title(f'Churn Rate by {col}')
    plt.xlabel(col)
    plt.ylabel('Number of Customers')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # Calculate churn rate for each category
    churn_rates = df_cleaned.groupby(col)['Churn'].mean().reset_index()
    churn_rates['Churn Rate (%)'] = churn_rates['Churn'] * 100
    print(f"\nChurn rates for '{col}':")
    display(churn_rates.sort_values(by='Churn', ascending=False))

# Numerical Features Analysis
numerical_cols = df_cleaned.select_dtypes(include=np.number).columns.tolist()
# Exclude 'Churn' itself from numerical columns for EDA if it's there
if 'Churn' in numerical_cols:
    numerical_cols.remove('Churn')

print("\nAnalyzing numerical features against Churn:")
for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df_cleaned, x=col, hue='Churn', kde=True, multiple='stack', palette='rocket')
    plt.title(f'Distribution of {col} by Churn')
    plt.xlabel(col)
    plt.ylabel('Density / Count')
    plt.tight_layout()
    plt.show()

    # Box plot for numerical features vs Churn
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df_cleaned, x='Churn', y=col, palette='pastel')
    plt.title(f'{col} vs Churn')
    plt.xlabel('Churn (0: No, 1: Yes)')
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()

# Correlation Matrix for numerical features
print("\nCorrelation Matrix for Numerical Features:")
plt.figure(figsize=(10, 8))
sns.heatmap(df_cleaned[numerical_cols + ['Churn']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features and Churn')
plt.show()

# Pairplot for selected numerical features (can be slow for many features)
# sns.pairplot(df_cleaned[['tenure', 'MonthlyCharges', 'TotalCharges', 'Churn']], hue='Churn', palette='viridis')
# plt.suptitle('Pairplot of Selected Numerical Features by Churn', y=1.02)
# plt.show()

print("\nEDA Complete. Proceed to feature engineering in the next notebook.")
```
